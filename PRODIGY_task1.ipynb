# Data Visualization

```python
years = [1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020 , 2023]

# Loop over each year and create the plot
for year in years:
    # Sorting and preparing data for the specific year
    total_population_sorted = df.sort_values(by=str(year), ascending=False)
    total_population_sorted['Population_Display'] = total_population_sorted[str(year)].apply(lambda x: f'{x / 1_000_000:.2f}M')
    total_top_ten_countries = total_population_sorted.head(10)

    # Print the top ten countries and their population for the current year
    print(f"Top ten countries of total population ({year})\n")
    print(total_top_ten_countries[['Country Name', str(year)]])

    # Bar plot for the current year
    plt.figure(figsize=(10, 6))
    sns.barplot(x=str(year), y="Country Name", data=total_top_ten_countries, palette="viridis")
    plt.title(f"Top Ten Countries by Population ({year})", fontsize=14)
    plt.xlabel("Population (in millions)", fontsize=12)
    plt.ylabel("Country Name", fontsize=12)
    plt.show()
```

```python
# List of years to analyze
years = [1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020, 2023]

# Loop over each year and create the plot for bottom 10 countries
for year in years:
    # Sorting and preparing data for the specific year
    total_population_sorted = df.sort_values(by=str(year), ascending=False)  # Sort in ascending order for bottom 10
    total_population_sorted['Population_Display'] = total_population_sorted[str(year)].apply(lambda x: f'{x / 1_000_000:.2f}M')
    total_bottom_ten_countries = total_population_sorted.tail(10)  # Select bottom 10

    # Print the bottom ten countries and their population for the current year
    print(f"Bottom ten countries of total population ({year})\n")
    print(total_bottom_ten_countries[['Country Name', str(year)]])

    # Bar plot for the current year (bottom 10 countries)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=str(year), y="Country Name", data=total_bottom_ten_countries, palette="coolwarm")
    plt.title(f"Bottom Ten Countries by Population ({year})", fontsize=14)
    plt.xlabel("Population (in millions)", fontsize=12)
    plt.ylabel("Country Name", fontsize=12)
    plt.show()
```

```python
# years list as mentioned
years = [1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020, 2023]

# Define a color palette
colors = sns.color_palette("tab20", len(years))

# Loop over each year and plot the histogram
for i, year in enumerate(years):
    plt.figure(figsize=(10, 6))
    sns.histplot(df[str(year)], kde=True, bins=30, color=colors[i])
    plt.title(f"Distribution of Population in {year} for All Countries", fontsize=14)
    plt.xlabel("Population", fontsize=12)
    plt.ylabel("Frequency", fontsize=12)

    # Format the x-axis to show population in millions
    plt.gca().set_xticklabels([f'{x / 1_000_000:.2f}M' for x in plt.gca().get_xticks()])

    plt.tight_layout()
    plt.show()
```

```python
# For top 10 countries (sort in descending order by 2023 population)
df_top_10 = df.sort_values(by='2023', ascending=False).head(10)

# For bottom 10 countries (sort in ascending order by 2023 population)
df_bottom_10 = df.sort_values(by='2023', ascending=True).head(10)

# Plot for Top 10 Countries
df_top_10_filtered = df_top_10[['Country Name', '1960', '2020', '2023']]
df_top_10_filtered.set_index('Country Name', inplace=True)
df_top_10_filtered.plot(kind='bar', figsize=(12, 7), width=0.8, colormap='Blues')
plt.title("Top 10 Countries by Population for Selected Years", fontsize=14)
plt.xlabel("Country Name", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Plot for Bottom 10 Countries
df_bottom_10_filtered = df_bottom_10[['Country Name', '1960', '2020', '2023']]
df_bottom_10_filtered.set_index('Country Name', inplace=True)
df_bottom_10_filtered.plot(kind='bar', figsize=(12, 7), width=0.8, colormap='Oranges')
plt.title("Bottom 10 Countries by Population for Selected Years", fontsize=14)
plt.xlabel("Country Name", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Population Distribution by Binned Categories (1960)
bins = [0, 10_000_000, 50_000_000, 100_000_000, 500_000_000, 1_000_000_000]
labels = ['0-10M', '10M-50M', '50M-100M', '100M-500M', '500M-1B']
df['Population_Binned'] = pd.cut(df['1960'], bins=bins, labels=labels)

population_bins = df['Population_Binned'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
population_bins.plot(kind='bar', color='palegreen')
plt.title("Population Distribution by Binned Categories (1960)", fontsize=14)
plt.xlabel("Population Range", fontsize=12)
plt.ylabel("Number of Countries", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
#Population Distribution by Binned Categories (2023)
bins = [0, 10_000_000, 50_000_000, 100_000_000, 500_000_000, 1_000_000_000]
labels = ['0-10M', '10M-50M', '50M-100M', '100M-500M', '500M-1B']
df['Population_Binned'] = pd.cut(df['2023'], bins=bins, labels=labels)

population_bins = df['Population_Binned'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
population_bins.plot(kind='bar', color='pink')
plt.title("Population Distribution by Binned Categories (2023)", fontsize=14)
plt.xlabel("Population Range", fontsize=12)
plt.ylabel("Number of Countries", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
#Top 10 Countries by Population Change (1960 to 2023)
df['Population_Change'] = df['2023'] - df['1960']  # Change from 1960 to 2023
df_sorted = df.sort_values(by='Population_Change', ascending=False)
top_change_countries = df_sorted[['Country Name', 'Population_Change']].head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='Population_Change', y='Country Name', data=top_change_countries, palette='pastel')
plt.title("Top 10 Countries by Population Change (1960 to 2023)", fontsize=14)
plt.xlabel("Population Change", fontsize=12)
plt.ylabel("Country Name", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Calculate the population growth rate for each country
df['Growth_Rate'] = ((df['2023'] - df['1960']) / df['1960']) * 100

# Sort the countries by growth rate
df_growth_rate = df[['Country Name', 'Growth_Rate']].sort_values(by='Growth_Rate', ascending=False).head(10)

# Create a bar plot to visualize the top 10 countries with the highest growth rates
df_growth_rate.set_index('Country Name', inplace=True)
df_growth_rate.plot(kind='barh', figsize=(12, 7), color='seagreen', legend=False)

plt.title("Top 10 Countries by Population Growth Rate (1960-2023)", fontsize=14)
plt.xlabel("Growth Rate (%)", fontsize=12)
plt.ylabel("Country Name", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
df_growth_rate_bottom = df[['Country Name', 'Growth_Rate']].sort_values(by='Growth_Rate', ascending=True).head(10)
df_growth_rate_bottom.set_index('Country Name', inplace=True)
df_growth_rate_bottom.plot(kind='barh', figsize=(12, 7), color='tomato', legend=False)

plt.title("Bottom 10 Countries by Population Growth Rate (1960-2023)", fontsize=14)
plt.xlabel("Growth Rate (%)", fontsize=12)
plt.ylabel("Country Name", fontsize=12)
plt.tight_layout()
plt.show()
```
